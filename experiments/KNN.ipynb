```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X = np.random.uniform(0,10,100)
sinX = np.sin(X) + np.random.normal(0,0.2,100)
plt.scatter(X,sinX)
```

```python
idx = np.argsort(X)
plt.scatter(X[idx],sinX[idx])
```

```python
def running_mean(x_data, y_data, window, online = True):
    
    idx = np.argsort(x_data)
    sorted_y_data = np.array(y_data[idx])
    
    means = []
    
    if online:
        means.append(sorted_y_data[:window].mean())
        for i in range(len(sorted_y_data)-window-1):
            loss = sorted_y_data[i]
            gain = sorted_y_data[i+window]
            means.append(means[-1] - loss/window + gain/window)
    else:
        for i in range(0,len(sorted_y_data)-window):
            data_window = sorted_y_data[i:i+window]
            means.append(data_window.mean())
            
    return means
```

```python
window = 5
means = running_mean(X, sinX,window,True)
plt.scatter(X,sinX)
plt.plot(sorted(X[window:]),means, color= 'r')
```

```python
window = 1
means = running_mean(X, sinX,window,True)
plt.scatter(X,sinX)
plt.plot(sorted(X[window:]),means, color= 'r')
```

```python
X = np.random.normal(0,1,(2,500))
X = np.concatenate([X,[[1]*500]],0)

Y = np.random.normal(3,1,(2,500))
Y = np.concatenate([Y,[[2]*500]],0)

Z = np.random.normal(-5,2,(2,500))
Z = np.concatenate([Z,[[3]*500]],0)




plt.scatter(X[0],X[1])
plt.scatter(Y[0],Y[1])
plt.scatter(Z[0],Z[1])
```

```python
dataset = np.concatenate([X.T,Y.T,Z.T])
plt.scatter(dataset[:,0],dataset[:,1])
```

```python
df = pd.DataFrame(dataset,columns = ['x','y','label'])
```

```python
def euclidean(x,y):
    if isinstance(x,np.ndarray) and isinstance(y,np.ndarray):
        return np.sqrt(((x - y)**2).sum(1))
    else:
        raise TypeError('Wrong types, should be numpy arrays')

def manhattan(x,y):
    if isinstance(x,np.ndarray) and isinstance(y,np.ndarray):
        return np.array([abs(xx - yy) for xx,yy in zip(x,y)]).sum()
    else:
        raise TypeError('Wrong types, should be numpy arrays')
```

```python
def knn_classifier(data, target, k, new_points, diff):
    res = []
    for points in new_points:
        distances = diff(np.array(points),data)
        res.append(np.argsort(distances)[:k])
    res = np.array(res)
    new_labels = target[res]
    return np.array([mode(np.array(nl))[0][0] for nl in new_labels])
```

```python
data = df[['x','y']].values
target = df['label'].values
new_points = [[0.3,2.5],[0,1],[3,4],[4,1.1],[-2.3,4],[-10,-10]]
new_labels = knn_classifier(data, target, 5, new_points, euclidean)
```

```python
plt.scatter(X[0],X[1],color = 'r')
plt.scatter(Y[0],Y[1],color = 'g')
plt.scatter(Z[0],Z[1],color = 'b')

points = np.array(new_points).T
plt.scatter(points[0],points[1],marker = 'X', color = 'black')
print(new_labels)
```

```python
X = np.random.uniform(0,10,100)
Y = np.random.uniform(1,5,100)
sinX = np.sin(X)+np.sin(Y)
plt.scatter(X,sinX)
```

```python
dataset = pd.DataFrame(np.array([X,Y, sinX]).T, columns = ['x','y','target'])
```

```python
data = dataset[['x','y']].values
target = dataset['target'].values
```

```python
def knn_regression(data, target, k, new_points, diff):
    res = []
    for points in new_points:
        distances = diff(np.array(points),data)
        res.append(np.argsort(distances)[:k])
    res = np.array(res)
    results = target[res]
    return np.array([nl.mean() for nl in results])




new_points = [[1,2],[3,4]]
diff = euclidean
k = 5
knn_regression(data, target, k, new_points, diff)
```

```python
xx = np.linspace(0,10,100)
yy = np.linspace(0,10,100)
newpoints = np.array([xx,yy]).T
diff = euclidean
k = 5
results = knn_regression(data, target, k, newpoints, diff)

sinX = np.sin(X)+np.sin(Y)
plt.scatter(X,sinX)

plt.scatter(xx,results)
```
