```python
import ML.mixtureModels as mm
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline  
```

# Generating data:

```python
x = np.random.normal(0,1,(2,100))
z = np.random.normal(0,1,(2,100))
corr = 0.4
L = np.array([[1,0],[corr,np.sqrt(1-corr**2)]])

m1 = [4,4]
m2 = [0,0]

m1 = np.array(m1)
m2 = np.array(m2)

y1 = m1.reshape(2,1) + np.dot(L,x)
y2 = m2.reshape(2,1) + np.dot(L,z)

fig = plt.figure(figsize=(8, 16))
ax1 = fig.add_subplot(211)
ax1.scatter(y1[0],y1[1])
ax1.scatter(y2[0],y2[1])
z = np.concatenate((y1,y2), axis = 1)

```

```python
fig = plt.figure(figsize=(8, 16))
ax1 = fig.add_subplot(211)
ax1.scatter(z[0],z[1],color = 'grey')

#set up initial centroids
mu1 = np.array([0,-3])
mu2 = np.array([0.2,8])

ax1.scatter(mu1[0],mu1[1],marker = 'X',c = 'red')
ax1.scatter(mu2[0],mu2[1],marker = 'X',c = 'red')

data = z.T
```

# Use kmeans model

```python
km = mm.kmeans()
km.setUp(data,2,np.array([mu1,mu2]))
km.fit()
```

```python
mu1 = km.dataSplit[0]['centroid']
mu2 = km.dataSplit[1]['centroid']
s1 = np.array(km.dataSplit[0]['datapoints']).T
s2 = np.array(km.dataSplit[1]['datapoints']).T
```

```python
fig = plt.figure(figsize=(8, 16))
ax1 = fig.add_subplot(211)
ax1.scatter(s1[0],s1[1])
ax1.scatter(s2[0],s2[1])
ax1.scatter(mu1[0],mu1[1],marker = 'X',c = 'red')
ax1.scatter(mu2[0],mu2[1],marker = 'X',c = 'red')
```
